### Text to Speech conversion using GTTS

gTTS (Google Text-to-Speech) is a Python library and CLI tool that converts text into spoken audio using Google's Text-to-Speech API. It supports multiple languages and accents and can save the output as an audio file or play it directly.

```python
#!pip install gtts
```

```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''gTTS (Google Text-to-Speech) is a Python library 
                      and CLI tool that converts text into spoken audio 
                      using Google's Text-to-Speech API. It supports multiple languages
                       and accents and can save the output as an audio file or play it directly.''',
                       lang='hi',tld='com')
text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file,autoplay=False)
```

### Text to Speech conversion using pyttsx3

#### Changing the gender using pyttsx3

pyttsx3 is an offline text-to-speech (TTS) conversion library in Python that works with different speech engines like sapi5 (Windows), NSSpeechSynthesizer (Mac), and espeak (Linux). Unlike gtts, it does not require an internet connection since it uses system-installed voices.

```python
#!pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio

text='''pyttsx3 is an offline text-to-speech (TTS) conversion library in Python that works
 with different speech engines like sapi5 (Windows), NSSpeechSynthesizer (Mac), 
 and espeak (Linux). Unlike gtts, it does not require an internet connection 
 since it uses system-installed voices.'''


audio = pyttsx3.init()
audio.setProperty('rate',150)
audio.setProperty('volume',0.8)

# change the voice
voice = audio.getProperty('voices')

# 0 for male and 1 for female

audio.setProperty('voice',voice[0].id) # for male voice

#audio.setProperty('voice',voice[1].id) # for female voice

# text to speech conversion

audio.say(text)

# save the audio file
audio.save_to_file(text, 'test_male_Voice.mp3')
#audio.save_to_file(text, 'test_female_Voice.mp3')

audio.runAndWait()
```
