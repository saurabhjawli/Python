```python
df[['HouseNO', 'Locality', 'City', 'PIN']] = df['ADDR'].str.extract(r'^(.*?),\s*(.*?),\s*(.*?)(?:,?\s*(\d{6}))?$')

'''
^: Matches the start of the string.
(.*?): Captures any characters lazily (up to the first comma ,). This corresponds to 'HouseNO'.
,\s*: Matches a comma followed by optional whitespace.
(.*?): Captures the next portion of the string (up to the next comma). This corresponds to 'Locality'.
,\s*: Matches another comma followed by optional whitespace.
(.*?): Captures the next portion (up to the end of the string or optional PIN). This corresponds to 'City'.
(?:,?\s*(\d{6}))?: Matches an optional block:
,?: Optionally matches a comma.
\s*: Matches optional whitespace.
(\d{6}): Captures exactly six digits (a PIN code). This corresponds to 'PIN'.
$: Matches the end of the string.
Result of str.extract: This method returns a DataFrame with each capture group ((...)) as a column.'''
```
